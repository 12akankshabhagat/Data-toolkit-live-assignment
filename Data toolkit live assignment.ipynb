## 1. Demonstrate three different methods for creating identical 2D arrays in NumPy. Provide the code for each method and the final output after each method

```python
## there are 3 method for creating identical 2D arrays in numpy.
# 1. np.array , 2. np.zeros , np.vstack .

# Method 1: Using np.array
array_1 = np.array([[1, 2, 3], [4, 5, 6]])
print(array_1)

# Method 2: Using np.zeros 
array_2 = np.zeros((2, 3), dtype=int)
array_2[0, :] = [1, 2, 3]
array_2[1, :] = [4, 5, 6]
print(array_2)

# Method 4: Using np.vstack
array_4 = np.vstack(([1, 2, 3], [4, 5, 6]))
print(array_4)
```

## 3.Explain the following terms: 1.The difference in np.array, np.asarray and np.asanyarray 2.The difference between Deep copy and shallow copy

```python
'''## 1.The difference in np.array, np.asarray and np.asanyarray 

# 1.np.array : ## its create object of original array and does not reflect any changes made to the original array.
Creates a new array from the input data, copying the data by default.
You can use the copy parameter to control whether a copy is made (copy=True by default).
Suitable for creating a new array with specified attributes (e.g., data type, order).

# 2.np.asarray: ## it would reflect all changes made to original array , convert in array which are not in array.
Converts input to an ndarray but avoids copying if the input is already an ndarray.
Useful for ensuring that an object is an ndarray without unnecessary data duplication.

# 3.np.asanyarray: ## convert input data to a  ndarray but does not copy if input already ndarray
Similar to np.asarray, but it will pass through subclasses of ndarray.
Useful if you want to preserve array subclasses (e.g., matrix or masked arrays).


## 2. The difference between Deep copy and shallow copy
# Deep Copy:
Creates a new top-level object and recursively copies all nested objects.
Nested objects are fully copied.
Changes to nested objects do not affect the original object.

# Shallow Copy:
Only creates a new top-level object.
Nested objects are not copied; they are referenced.
Changes to nested objects affect both the original and the copied object. '''
```

## 14.Create a time-series dataset in a Pandas DataFrame with columns: 'Date', 'Temperature', 'Humidity' and Perform the following tasks using Matplotlib: a) Plot the 'Temperature' and 'Humidity' on the same plot with different y-axes (left y-axis for 'Temperature' and  right y-axis for 'Humidity'). b) Label the x-axis as 'Date'.c) Set the title of the plot as 'Temperature and Humidity Over Time'.    

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

date_range = pd.date_range(start='2023-01-01', periods=100, freq='D')
temperature = np.random.uniform(15, 30, size=100)
humidity = np.random.uniform(30, 90, size=100)

df = pd.DataFrame({
    'Date': date_range,
    'Temperature': temperature,
    'Humidity': humidity
})

fig, ax1 = plt.subplots()

ax1.plot(df['Date'], df['Temperature'], color='red', label='Temperature')
ax1.set_xlabel('Date')
ax1.set_ylabel('Temperature (°C)', color='red')
ax1.tick_params(axis='y', labelcolor='red')

ax2 = ax1.twinx()
ax2.plot(df['Date'], df['Humidity'], color='blue', label='Humidity')
ax2.set_ylabel('Humidity (%)', color='blue')
ax2.tick_params(axis='y', labelcolor='blue')

# Set the title of the plot
plt.title('Temperature and Humidity Over Time')

# Show the plot
fig.tight_layout()
plt.show()
```

## Q.15 Create a NumPy array data containing 1000 samples from a normal distribution. Perform the following tasks using Matplotlib:  a) Plot a histogram of the data with 30 bins  b) Overlay a line plot representing the normal distribution's probability density function (PDF)  c) Label the x-axis as 'Value' and the y-axis as 'Frequency/Probability   d) Set the title of the plot as 'Histogram with PDF Overlay 

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
data = np.random.normal(loc=0, scale=1, size=1000)
plt.hist(data, bins=30, density=True, alpha=0.6, color='g', label='Histogram')
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = norm.pdf(x, np.mean(data), np.std(data))

plt.plot(x, p, 'k', linewidth=2, label='PDF')

plt.xlabel('Value')
plt.ylabel('Frequency/Probability')
plt.title('Histogram with PDF Overlay')
plt.legend()
plt.show()
```

## 17.Create a Seaborn scatter plot of two random arrays, color points based on their position relative to the origin (quadrants), add a legend, label the axes, and set the title as 'Quadrant-wise Scatter Plot'.

```python
# Generate two random arrays
x = np.random.randn(100)
y = np.random.randn(100)

# Create a DataFrame
df = pd.DataFrame({'x': x, 'y': y})

# Define function to determine quadrant
def get_quadrant(x, y):
    if x > 0 and y > 0:
        return 'Q1'
    elif x < 0 and y > 0:
        return 'Q2'
    elif x < 0 and y < 0:
        return 'Q3'
    else:
        return 'Q4'

# Apply the function to determine the quadrant for each point
df['Quadrant'] = df.apply(lambda row: get_quadrant(row['x'], row['y']), axis=1)

# Create the scatter plot using Seaborn
plt.figure(figsize=(10, 6))
scatter_plot = sns.scatterplot(data=df, x='x', y='y', hue='Quadrant', palette='deep', style='Quadrant', s=100)

# Label the axes
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

# Set the title
plt.title('Quadrant-wise Scatter Plot')

# Add a legend
plt.legend(title='Quadrant')

# Show the plot
plt.show()
```

## 18.With Bokeh, plot a line chart of a sine wave function, add grid lines, label the axes, and set the title as 'Sine Wave Function

```python
import numpy as np
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
x = np.linspace(0, 4 * np.pi, 100)
y = np.sin(x)

output_notebook()

p = figure(title="Sine Wave Function", x_axis_label='X-axis', y_axis_label='Y-axis')
p.line(x, y, legend_label="Sine Wave", line_width=2)
show(p)
```

## 19.Using Bokeh, generate a bar chart of randomly generated categorical data, color bars based on their values, add hover tooltips to display exact values, label the axes, and set the title as 'Random Categorical Bar Chart

```python
import numpy as np
from bokeh.plotting import figure, show, output_notebook
from bokeh.io import output_notebook
from bokeh.models import ColumnDataSource, HoverTool
from bokeh.transform import factor_cmap
from bokeh.palettes import Spectral6

# Enable output to notebook
output_notebook()

# Define categories and generate random values
categories = ['A', 'B', 'C', 'D', 'E']
values = np.random.randint(1, 100, size=len(categories))

# Create ColumnDataSource
source = ColumnDataSource(data=dict(categories=categories, values=values))

# Create a figure
p = figure(x_range=categories, title="Random Categorical Bar Chart", 
           x_axis_label='Categories', y_axis_label='Values')

# Add vertical bars to the plot
p.vbar(x='categories', top='values', width=0.9, source=source, 
       legend_field="categories", line_color='white', 
       fill_color=factor_cmap('categories', palette=Spectral6, factors=categories))

# Add hover tool
hover = HoverTool(tooltips=[("Category", "@categories"), ("Value", "@values")])
p.add_tools(hover)

# Customize the legend
p.legend.title = 'Categories'
p.legend.label_text_font_size = '8pt'

# Display the plot
show(p)
```
